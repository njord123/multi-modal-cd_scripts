```python
from sklearn.metrics import classification_report
def conf_matrix(pred_arr, actu_arr, labels=None):
    y_pred = pd.Series(pred_arr, name='Predicted')
    y_actu = pd.Series(actu_arr, name='Actual')
    # print(classification_report(y_actu, y_pred))
    df_confusion = pd.crosstab(y_actu, y_pred, rownames=['Actual'], colnames=['Predicted'], normalize='index')
    display(df_confusion)

    ax = plt.subplot()
    sns.heatmap(df_confusion, annot=True, ax=ax)
    if labels != None:
        ax.xaxis.set_ticklabels(labels)
        ax.yaxis.set_ticklabels(labels)
    plt.show()

    return df_confusion
```
